# Caso Real de Big Data: Telemetría y Analítica en la Formula 1

### 1. ¿Que datos genera?
Un monoplaza de Formula 1 es basicamente una computadora sobre ruedas que genera datos masivos de telemetría.
* **Tipo de datos:** Datos estructurados de sensores (temperatura de neumaticos, presion de frenos, revoluciones del motor, fuerza G, desgaste aerodinamico) y datos no estructurados (comunicaciones por radio, video, clima).
* **Volumen:** Cada auto tiene mas de 300 sensores. En un solo fin de semana de carrera, los dos autos de un equipo generan mas de 2 a 3 Terabytes de datos.
* **Frecuencia:** Los datos son de alta frecuencia y en **tiempo real**. Se transmiten mas de 1.1 millones de puntos de datos por segundo desde el auto hacia los servidores del equipo.

### 2. ¿Quien los usa?
* **Ingenieros de pista (en el muro de boxes):** Consumen los datos en tiempo real para tomar decisiones inmediatas sobre la carrera.
* **Cientificos de datos (en la fabrica):** Equipos completos ubicados en la sede central analizan los datos en paralelo para hacer simulaciones.
* **Los pilotos:** Reciben instrucciones basadas en los datos para cambiar su estilo de conduccion.
* **La transmision de TV:** Usan estos datos para mostrar graficos predictivos a los espectadores.

## 3. ¿Que problema resuelve?
* **Ventaja competitiva:** En la F1, ganar o perder se decide por milesimas de segundo. El analisis de datos permite ajustar la aerodinamica y la estrategia de paradas en boxes para vencer al rival.
* **Mantenimiento predictivo:** Ahorra millones de dolares. El sistema analiza patrones y avisa si una pieza del motor esta a punto de fallar aantes de que se rompa, evitando que el auto abandone la carrera.
* **Seguridad:** Los sensores detectan anomalias microscopicas permitiendo llamar al piloto a boxes antes de sufrir un accidente grave.

### 4. ¿Por que no bastaria Excel?
Intentar usar una hoja de calculo o una base de datos relacional tradicional en la F1 seria imposible debido a diferentes caracteristicas como:
* **Velocidad de ingestion:** Excel no esta diseñado para recibir un flujo de streaming de mas de 1 millon de registros por segundo. El programa colapsaria instantaneamente.
* **Limite de Volumen:** Excel tiene un limite maximo de 1,048,576 filas. Un solo auto llenaria un archivo entero de Excel en menos de 1 segundo de carrera.
* **Complejidad Analitica:** Los equipos utilizan estos datos para ejecutar miles de "Simulaciones de Monte Carlo" por segundo y entrenar modelos de Machine Learning. Una hoja de calculo no tiene la arquitectura ni el poder de procesamiento paralelo para predecir escenarios probabilisticos en microsegundos.

### Justificación
El caso de la telemetria en la Formula 1 se clasifica indudablemente como Big Data porque ilustra a la perfeccion las tres dimensiones fundamentales de este concepto, conocidas como las "3 V". Para empezar, destaca por su inmenso **volumen**, ya que un solo equipo puede generar entre 2 y 3 Terabytes de informacion en apenas un fin de semana de carrera, superando ampliamente las capacidades del almacenamiento tradicional. A esto se le suma la extrema **velocidad** con la que opera el ecosistema; la informacion no se guarda para un analisis futuro, sino que se transmite y procesa en tiempo real, emitiendo mas de 1.1 millones de puntos de datos por segundo mientras los vehiculos compiten a mas de 300 kh/m. Finalmente, el caso ejemplifica una gran **variedad** de informacion, integrando simultaneamente datos estructurados numericos (como temperaturas, metricas de desgaste y presion de sensores) con datos no estructurados, como comunicaciones de audio por radio, transmisiones de video en vivo de las camaras on-board y radares meteorologicos dinamicos.
